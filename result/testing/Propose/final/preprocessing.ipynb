```python
import pandas as pd
```

```python
group_0 = pd.read_csv('./group_0.csv', header=None)
group_1 = pd.read_csv('./group_1.csv', header=None)
group_2 = pd.read_csv('./group_2.csv', header=None)
len(group_0), len(group_1), len(group_2)
```

```python
classifier_v2 = pd.read_csv('./t2_classifier_v2-epoch_4.csv', header=None)
classifier_v3 = pd.read_csv('./t2_classifier_v3-epoch_7.csv', header=None)
len(classifier_v2), len(classifier_v3)
```

## Use one

# Merge and Multiply

```python
def multiply(row: pd.Series):
    id = row[0]
    labels = row[-10:]
    assert len(labels) == 10, len(labels)
    pred = []
    for i in range(10):
        pred.append(row[1] * row[4 + i] + row[2] * row[14 + i] + row[3] * row[24 + i])
    return pd.Series([id, *pred, *labels])
```

## Version 2

```python
summary_v2 = classifier_v2.iloc[:, :4]
summary_v2 = summary_v2.merge(group_0.iloc[:, :11], on=0)
summary_v2 = summary_v2.merge(group_1.iloc[:, :11], on=0)
summary_v2 = summary_v2.merge(group_2, on=0)
```

```python
summary_v2.columns = range(44)
pred_data_v2 = summary_v2.apply(multiply, axis=1)
```

```python
pred_data_v2 = pred_data_v2.astype({0: int})
pred_data_v2.to_csv('./classifier_merge_v2.csv', header=None, index=None)
```

## Use one

```python
def choose_best(row: pd.Series):
    id = row[0]
    labels = row[-10:]
    assert len(labels) == 10, len(labels)
    maxidx = row[1:4].idxmax() - 1
    pred = row[4 + maxidx * 10:4 + maxidx * 10 + 10]
    return pd.Series([id, *pred, *labels])
```

### Version 2

```python
use_best_v2 = summary_v2.apply(choose_best, axis=1)
```
