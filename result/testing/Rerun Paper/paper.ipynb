```python
import utils
import pandas as pd
```

```python
metrics = {'euclid': utils.euclid, 'emd_r1': utils.emd, 'emd_r2': lambda x, y: utils.emd(x, y, r=2)}
corrs = {'lcc': utils.lcc, 'srcc': utils.srcc}
```

```python
def sub_test(file, start = None, end = None, title = ''):
    data = pd.read_csv(file, header=None)
    # data = preds.merge(pd.read_csv('./merge__test_labels.csv', header=None), on=0)
    sum_data = utils.summarize(data=data, metrics=metrics, corrs=corrs)
    filter = utils.show_detail(sum_data, 'mean', figsize=(20, 10), count=min(len(sum_data), 2000), title=title, start=start, end=end)
    utils.compare_with_ground_truth(filter, ['mean', 'std'])
    return sum_data
```
